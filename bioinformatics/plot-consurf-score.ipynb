```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = np.loadtxt('consurf-scores-normalized.txt')
y = np.arange(len(l))
y += 27
```

```python
plt.figure(figsize=(4,15))
plt.plot(l,y,'g')
ylim = plt.ylim(0,1146)
plt.gca().invert_yaxis()
plt.axvline(x=0,c='k', ls='--')
#plt.axhline(y=324,c='b',ls='-',lw=3)
#plt.axhline(y=518,c='b',ls='-',lw=3)
plt.axhspan(324,518,alpha=0.5,color='cyan')
plt.text(4,450,'RBD',fontsize=20)
plt.ylabel('Residue index',fontsize=24)
plt.xlabel('ConSurf score',fontsize=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.savefig('consurf-score.png',dpi=300,bbox_inches='tight')
```

```python
cc = 27 #buffer

ntd = np.mean(l[0:324-cc])
print(ntd)

rbd = np.mean(l[324-cc:518-cc])
print(rbd)

linker = np.mean(l[518-cc:685-cc])
print(linker)

s2 = np.mean(l[685-cc:-1])
print(s2)
```

```python
mean_values = [s2,linker,rbd,ntd]
labels = ['S2','Linker','RBD','NTD']
colors = ['yellow','red','green','cyan']



fig, ax = plt.subplots(figsize=(4,15))
#ax.tick_params(axis="y",direction="in", pad=-90)
ax.barh(labels,mean_values, color=colors)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel('Mean Consurf Score',fontsize=24)
plt.axvline(x=0,color='k',ls='--')
plt.xlim(-0.3,0.1)
plt.savefig('mean-consurf-score.png',dpi=300,bbox_inches='tight')
```

```python
#print high mutable residues
cc=27
f1 = open('high-consurf-scores.txt','w')
print('#residue number  #consurf score',file=f1)
for i in range(len(l)):
    if l[i] > 2.0:
        print(i+cc,l[i],file=f1)
f1.close()
```
