```python
diff_1 = diff_part['diff']
diff_2 = diff_full['diff']
```

```python
diff_1_rbd = diff_1[323:518]
diff_2_rbd = diff_2[323:518]
```

```python
diff_1_rbd_avg = np.linalg.norm(diff_1_rbd, axis = 0)
x = np.arange(len(diff_1_rbd_avg))
plt.figure(figsize=(20,10))
plt.plot(x,diff_1_rbd_avg,'b',lw=0.5)
plt.xlabel('Residue index',fontsize=18)
plt.ylabel('RMS $\Delta$ Correlation',fontsize=18)
plt.title('norm(Partially open state - Closed state); RBD only',fontsize=18)
xticks = [0,499,999,1499,1999,2499,2999,3437]
xlabels = np.asarray(xticks)+1
plt.xticks(ticks=xticks,labels=xlabels,fontsize=16)
plt.yticks(fontsize=16)
plt.xlim(0,3440)
plt.savefig('diff_partially_open_avg.png',dpi=300,bbox_inches='tight')
```

```python
print(diff_1_rbd_avg.shape)
```

```python
f1 = open('diff_partially_open_avg.txt','w')
print("#Chain     #Residue_index   #Correlation difference norm",file=f1)
for i in range(len(diff_1_rbd_avg)):
    if i < 1146:
        chain = 'A'
        resid = i+1
    elif i < 2292 :
        chain = 'B'
        resid = i+1 - 1146
    else :
        chain = 'C'
        resid = i+1 - 2292
    
    print(chain,resid,diff_1_rbd_avg[i],file=f1)
f1.close()
```
