Compute the difference between Linear Mutual Information based cross correlation between the closed and partially open structures

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
closed = np.loadtxt('closed_80ns/mutual_info.txt')
partially_open = np.loadtxt('partially_open_80ns/mutual_info.txt')

diff_part = np.subtract(partially_open,closed)
```

```python
diff_1 = diff_part
plt.figure(figsize=(33,5))
cb = plt.imshow(diff_1[323:518],cmap='RdBu',aspect='auto',vmax=0.5,vmin=-0.5)
plt.xlabel('Residue index',fontsize=20)
plt.ylabel('RBD residues',fontsize=20)
plt.title('$\Delta$ Correlation (Partially open state - Closed state)',fontsize=18)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ticks=[0,25,50,75,100,125,150,175,194]
labels=np.asarray(ticks)+324
plt.yticks(ticks=ticks,labels=labels,fontsize=20)
xticks = [0,499,999,1499,1999,2499,2999,3437]
xlabels = np.asarray(xticks)+1
plt.xticks(ticks=xticks,labels=xlabels)
cbar = plt.colorbar(cb,fraction=0.046, pad=0.04)
cbar.ax.tick_params(labelsize=18)
plt.savefig('diff_partially_open.png',dpi=100,bbox_inches='tight')
```

```python
fully_open = np.loadtxt('fully_open_80ns/mutual_info.txt')
diff_full = np.subtract(fully_open,closed)
```

```python
diff_2 = diff_full
plt.figure(figsize=(33,5))
cb = plt.imshow(diff_2[323:518],cmap='RdBu',aspect='auto',vmax=0.5,vmin=-0.5)
plt.xlabel('Residue index',fontsize=20)
plt.ylabel('RBD residues',fontsize=20)
plt.title('$\Delta$ Correlation (Fully open state - Closed state)',fontsize=18)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ticks=[0,25,50,75,100,125,150,175,194]
labels=np.asarray(ticks)+324
plt.yticks(ticks=ticks,labels=labels,fontsize=20)
xticks = [0,499,999,1499,1999,2499,2999,3437]
xlabels = np.asarray(xticks)+1
plt.xticks(ticks=xticks,labels=xlabels)
#ticks(fontsize=16)
cbar = plt.colorbar(cb,fraction=0.046, pad=0.04)
cbar.ax.tick_params(labelsize=18)
plt.savefig('diff_fully_open.png',dpi=100,bbox_inches='tight')
```
